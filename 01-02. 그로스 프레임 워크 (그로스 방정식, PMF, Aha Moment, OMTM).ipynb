## 그로스 방정식
- **사업 성장 (Growth)에 핵심적인 영향을 미치는 인과관계와 상관관계 변수들의 수식**
    - **그로스 퍼널을 이루는 고객 지표가 무수히 많아, 마케터가 집중해야 하는 핵심 지표를 정의**

$$
\text{사업 성과} = (\text{변수 1} + \text{변수 2}) \times \text{변수 3}
$$

<br>

### 그로스 방정식의 예시

<br>

#### 기대 매출 (1)

$$\text{기대 매출} = \{(\text{회원가입} \times  \text{전환율}) \times \text{LTR} + (\text{기존 회원}) \times \text{LTR}\}$$

#### 기대 매출 (2)

$$\text{기대 매출} = \{(\text{DAU} \times \text{전환율} \times \text{AOV}) \times \text{재구매율}\} \times (1 - \text{이탈율})$$

<br>

- **CASE 1**
    - 충분히 높은 DAU
    - AOV와 잔존율은 양호 추세
    - **단기적인 구매 전환율과 재구매율이 감소** 

    $\rightarrow$ **DAU의 성장세를 유지하면서 구매 전환율과 재구매율을 개선하기 위한 개선점 탐색**

- **CASE 2**
    - 구매 전환율, AOV, 재구매율은 성장 추세
    - DAU와 잔존율이 감소추세

    $\rightarrow$ **구매 전환율은 유지시키면서, DAU를 성장시킬 수 있는 개선점 탐색**


<br>

### 그로스 방정식 활용
- 그로스 방정식을 활용할 때
    - 사업과 서비스의 라이프 사이클에 따라 변화/발전
    - 밝혀지는 인과 / 상관관계를 바탕으로 지속적인 유지 보수 필요
    - 일에 따라 방정식을 바꾸는 것이 아니라 방정식에 따라 일을 바꿈


<br>

        - **상품 당 리뷰수** : 최근에 작성된 상품당 리뷰수의 영향을 받는다면, 상품당 최신 리뷰 수 확인 필요

## PMF (Product - Market Fit)
- 자사 제품이 제공하는 핵심 기능이나 가치가 고객의 핵심 니즈를 확실히 충족해 주고 있는 상태
- 막 시작하는 서비스나 그로스 마케팅을 처음 시도해보는 상황에서 PMF를 확인하는 것은 매우 중요

<br>

> 자사 제품이 제공하는 핵심 기능이나 가치가 고객의 핵심 니즈를 확실히 충족해 주고 있는 상태 = **사업의 지속 가능성이 확보된 상태**

<br>

### PMF 지표

<br>

#### 1. 리텐션
- **오늘 새롭게 방문한 고객이, 지속적으로 재방문하고 있는가?**


$$
\frac{\text{N일 뒤에도 방문하는 고객}}{\text{오늘 새롭게 방문한 고객}}
$$

<br>

#### 2. 전환율
- 오늘 방문한 고객 중 몇 명이 실제로 지갑까지 열었는가?

$$
\frac{\text{오늘 결제한 고객 수}}{\text{오늘 방문한 고객 수}}
$$

<br>

#### 3. NPS
- 제품을 사용해 본 고객 중, 적극적으로 추천하는 고객이 많은가?
- 설문이 필요한 지표

$$
\frac{\text{추천 고객 - 비추천 고객}}{\text{전체 응답자}}
$$

<br>

#### PMF 달성의 근거
1. **업계 평균 / 경쟁사 대비**

<br>

2. **CAC / LTV**
- LTV는 구매 전환율과 리텐션이 높아질수록 커짐

    $\rightarrow$ 현재의 CAC가 LTV보다 크다면, PMF를 달성하지 못한 것

<br>

3. **BEP 관점**
- **BEP : Break-Given Point, 손익 분기점**
- **현재 상황이 유지된다는 전제하에 BEP를 달성하기 위한 리텐션과 구매 전환율을 파악하여, PMF 달성의 기준으로 삼는 것**
- BEP는 고객 관점의 PMF가 아닌, 사업 기준의 PMF

<img src='img/0204.png' width=600>

<br>

### PMF의 중요성

<br>

#### PMF 진단의 유의점

- **CASE 1**: 신규 서비스가 런칭 후 1개월 만에 폭발적인 관심 $\rightarrow$ 활성 고객 1만명

    $\rightarrow$ 마케터는 PMF가 달성되었다고 판단

    $\rightarrow$ **이러한 관심의 지속성은 고민하지 않음**

<br>

- **CASE 2**: 초기의 반응을 PMF로 착각하여 대대적인 투자
    - 특별한 마케팅을 하지 않았는데도 급격한 활성 유저가 증가하였으니 대대적으로 마케팅을 시작하면 폭발적인 성장을 할 수 있다고 믿는 것

    $\rightarrow$ 초반에는 반짝하는 폭발적인 반응이 금방 식을 수 있음

    $\rightarrow$ **PMF 달성 여부가 확실히 확인되지 않았다면 아마 초반의 반짝 성과는 지속 가능하지 않음**

    $\rightarrow$ **서비스 런칭 초기시, 일반인들보다 호기심이 높은 소수의 집단이 서비스를 활용하는 경향이 있기 때문**


<img src='img/0205.png' width=700>

<br>

- 리텐션, 구매 전환율, NPS 점수를 산출해보며 서비스가 본격적인 마케팅을 할 준비가 되었는지 점검 필요
- PMF가 달성되지 않았다는 말은 제품의 지속 가능성이 확보되지 않았다는 의미
- **마케팅 예산의 투입은 PMF 달성 이후,**

    **PMF 달성되기 전이라고 판단된다면, 초기 유저가 떠나지 않도록 제품의 실질 가치 개선에 마케터의 노력을 집중해야 함**

<br>

#### AARRR의 올바른 순서
- 실제 그로스 마케터는 먼저 PMF 달성 여부를 확인하고

    $\rightarrow$ 그로스 방정식을 그려, 고객의 특별한 여정에서 어떤 지표에 집중해야 하는지 고민

    $\rightarrow$ 각 지표의 Input과 전환을 개선하기 위한 고민

## Aha Moment
- **제품의 핵심 가치가 고객에게 전달되는 구체적인 순간**

<img src='img/0207.png' width=700>

<br>

- Aha Moment를 경험한 고객 (서비스의 핵심 가치를 경험한 고객) 서비스에 만족하여, 오래남아 큰 가치를 만들어내는 고가치 고개긍로 변환
- Aha Moment를 경험하지 못한 고객들은 서비스에 특별함을 느끼지 못하고 금세 이탈

    $\rightarrow$ **Aha Moment를 찾아 그것을 더 많은 고객들이 경험하도록 만드는 것**

<br>

- Aha Moment를 경험한 고객들의 리텐션 그래프는 모양이 다르기 떄문에, 
    
    다양한 고객군들의 리텐션 그래프 중 모양이 다른 그래프를 찾아 Aha Moment를 정의 가능

<img src='img/0208.png' width=600>

<br>

### Aha Moment 탐색

<br>

#### 1. 세그먼트 탐색
- Retention Plateau가 만들어지는 세그먼트 탐색
    - **연역적** : 제품 기능의 경험 여부로 구분
    - **귀납적** : 제품 사용자 중 최상위/최하위 구분

<img src='img/0209.png' width=600>

<br>

#### 2. 세그먼트의 특징 탐색
- Retention Plateau가 만들어진 세그먼트의 특징 탐색
- 리텐션이 높은 그룹과 특정 액션의 교집합을 통해 리텐션이 높은 그룹의 고객들 중 어느 정도의 비율이 특정 액션을 했는지를 탐색

    $\rightarrow$ **교집합에 있는 고객의 행동이 곧 Aha Moment**

    $\rightarrow$ **교집합의 크기가 너무 크거나 작다면 분석에 경계를 해야함**

<br>

- **교집합 크기의 의미**
    - **교집합이 너무 큼 (50 ~ 60% 보다 높음)** : 너무 당연한 행동이기 떄문에 Aha Moment가 아닐 확률이 높음
    - **교집합이 너무 작음 (10 ~ 20%)** : 소수만 경험한 기능을 보고, 이 기능 때문에 세그먼트 고객들의 리텐션이 좋아졌다고 말하기 힘듬

### Aha Moment 적용

- **D와 E그룹의 리텐션 그래프에서 Plateau 형성**

    - 액션 1은 비율이나 빈도를 봤을 때, D와 E그룹에서 특별히 높은 수치 X
    - **액션 2는 A그룹에서 E그룹으로 갈수록 비율과 빈도가 증가하지만, 전체적으로 비율이 높음** 

        $\rightarrow$ **액션 2는 서비스 내에서 취하는 당연한 행동일 가능성**
    - **액션 3은 A ~ C 그룹 대비 D ~ E 그룹이 더 많이 하는 액션 $\rightarrow$ 액션 3은 Aha Moment에 해당되는 고객의 경험**

### Aha Moment 활용

<br>

- **1단계 : 더 많은 사용자들이 더 쉽고 깊게 잠정적인 Aha Moment를 접할 수 있도록 조치**

    $\rightarrow$ CRM, 광고, UI/UX 변경 등

- **2단계 : 해당 유저들의 리텐션이 높아지고 리텐션 그래프에 Plateau가 만들어 지는지 확인**

    $\rightarrow$ 일시적인 상승에 유의

- **3단계 : 안정적인 리테션과 Plateau가 유지된다면, Aha Moment의 전/후 퍼널의 Input을 집중**
