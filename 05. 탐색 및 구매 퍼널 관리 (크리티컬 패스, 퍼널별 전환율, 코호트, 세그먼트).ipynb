# 탐색 및 구매 퍼널 관리

<br>

### 탐색 및 구매 퍼널에는 정형화된 퍼널이 존재하지 않음
- 결제와 같은 아웃풋 행동을 중심으로 설게하는 것이 바람직

<br>

- **이커머스**
    - 탐색 $\rightarrow$ 가입 $\rightarrow$ 대안 비교 $\rightarrow$ 장바구니 $\rightarrow$ 결제

<img src='img/0701.png' width=800>

<br>

- **구독 서비스**
    - 가입 $\rightarrow$ 무료 사용 $\rightarrow$ 대안 비교 $\rightarrow$ 결제

<img src='img/0702.png' width=800>

<br>

- **가입이 '탐색'단계의 초반 허들이 되듯이, 결제 직전에는 대안 비교가 가장 큰 허들이 되는 경우가 빈번**
- **서비스나 제품별로 다른 고객들의 대안 비교 행태를 파악하고 이 저항을 없애줌으로써 결제 전환을 높일 수 있음**

| **주요 탐색/구매 퍼널** | **중요성** |
| - | - |
| 가입이 언제 어떤 방식으로 일어나는가? | 가입은 고객이 서비스에 대해서 최소한의 노력, 시간과 개인정보라는 비용을 투입해야 하는 1차적인 허들|
|무료 사용이 가능한가? | - 전/후로 탐색 활성도 상승<br>- 고객의 사용후 행동을 살펴보면서 고객이 무엇에 마족하고 불만족 하는지 유추 |
| 고객들의 대안 비교는 언제 어떻게 일어나는가? | - 결제라는 마지막 행동으로 이어지기 전에 대안 비교라는 것이 **가장 높고 강한 허들**<br>- 탐색 활성도가 높아져서 고객이 우리 서비스를 잘 알게 된다고 하더라도 대안 비교 과정에서 다른 더 좋은 대안이 있다고 판단하게 되면 최종 행동에 다다르지 못함<br>- 고객이 어느 시점에서 어떻게 대안 비교를 하는지 파악해서 그 단계에서 이탈이 예상되는 고객의 전환을 유도해야 함 |
| 결제는 어느 단계에서 이루어 지는가? | - 결제는 고객이 서비스에 지불하는 가장 높은 수준의 비용<br>- 고객이 구매에 대한 확신을 가지기 위, 고객이 결제를 통해 최대 효용을 얻기 위해 다시 활성도가 증가|

### 탐색 및 구매 퍼널
- **탐색 구매 퍼널에서 핵심적인 지표는 각 퍼널 단계별로 존재하는 전환율**

<img src='img/0704.png' width=600>

<br>

> **전환율** = 다음 액션에 대한 **필요성** - **저항** (불편함)

<br>

- 전환율을 높이기 위해서는 다음 단계에 기다리는 욕구를 키우고, 이 단계에서 느끼는 저항을 최소화

## 탐색 및 구매 퍼널 핵심 지표

<br>

### 구매 전환율
- 매일 결제가 일어나지 않는 종류의 서비스는 일단위로 DAU대비 결제자의 비율을 보는 것은 의미가 없을 수 있으며

    오히려 월 단위의 MAU대비 월 결제자의 비율을 보는 것이 더 의미가 있을 수도 있음

$$\frac{\text{결제한 유저의 수}}{\text{전체 유저의 수}}$$

<br>

### ARPU
- ARPU는 MAU나 DAU의 중요송을 과대 해석하게 된다는 부작용 존재

    MAU에서 구매하는 고객의 비율이 작을 수도 있기 때문

$$\frac{\text{전체 매출}}{\text{전체 유저의 수}}$$


<br>

### ARPPU

$$\frac{\text{전체 매출}}{\text{결제한 유저의 수}}$$

## 탐색

<br>

### 흔적과 시그널
- **탐색 과정은 다양한 행동을 포함**

1. **어떤 상품이나 서비스에 관심을 보였는가?**
- 시그널을 활용해 다양한 개인화를 통하여 구매를 유도

2. **어떤 상품이나 서비스에 불만족 했는가?**

<img src='img/0705.png' width=800>

<br>

- **탐색 퍼널에서 이탈의 징후를 파악했다면, 더 깊게 고객의 행동과 생각을 파악하여 예방하기 위한 활동이 필요**

    $\rightarrow$ 구매, 구매 이후의 재구매나 잔존의 퍼널로 넘어갈 수 있음

- 특히 무료 사용 이후나, 장바구니에 상품을 넣은 직후 이탈했다면 심도 깊은 이해가 필요 
    
    $\rightarrow$ **서비스의 핵심 가치를 경험했으면서도 결제 단계로 넘어가지 못했기 떄문 = 핵심 퍼널 이탈**

- 핵심 퍼널에서 이탈 : 대안 비교 과정에서 경쟁사와의 비교를 통해 경쟁사로 이탈하여, 그곳에서 결제가 이루어 졌거나, 더 훌륭한 대체제를 찾았음을 의미

    $\rightarrow$ **제품과 서비스의 경쟁력에 대한 근본적인 개선이 필요 (고객 조사 등이 필요)**

## 크리티컬 패스
- **크리티컬 패스 : 서비스로 진입하는 순간부터 아하 모먼트까지 도달하는 고객의 경로**
- **아하 모먼트** : 핵심 가치가 전달되어 고객의 반응률과 리텐션이 개선되는 핵심 액션
    - 리텐션 그래프에 플래토가 생긴 고객들을 모아 그들의 공통 행동을 분석해보면 아하 모먼트 탐색 가능

- **크리티컬 패스를 점검하기 위해서, 아하 모먼트에 이르는 기존의 크리티컬 패스를 그려야 함**
    - 고객의 이탈률이나 불편함이 큰 지점을 찾아 거시적인 흐름 차원에서 더 많은 고객들이 아하 모먼트를 빨리, 쉽게 경험할 수 있는지 탐색
- 각 서비스와 제품은 고유한 탐색 및 구매 퍼널을 가지기에, 고유한 탐색 구매 퍼널에서 주요 경로별 전환율을 파악하여 이탈률이 높은 지점을 찾고 개선

<br>

- **예) 여행 앱**
    - 아하 모먼트 : 여행 예정지의 베스트 투어 상품을 보여 주는 것

## 코호트
- 특정한 공통점을 공유하는 고객들의 집단을 **기간 경과**에 따라 추적하는 것
    - 세그먼트에는 시간 경과에 따른 추적의 개념이 포함되어 있지 않음

<br>

### 코호트 활용


<br>

- **충성 고객의 행동을 일반 고객에 유도**
    - 가장 일반적인 방법
    - 충성 고객의 탐색 활동을 입란 고객도 많이 하도록 유도하면 탐색율 증가
- **유입 경로 별 고객 행동 차이 파악**
    - 유입 경로에 따른 진입 고객들의 차이점을 파악하고, 전한에 유리한 고객들을 더 많이 유입시켜 전환율 개선

- **내부 데이터를 중심으로 고객 행동 분석 후, 고객 설문이나 인터뷰 통해서 행동 동기 파악**

<img src='img/0711.png' width=800>

<br>

- 코호트 예시
    - 일반 고객과 충성 고객은 원하는 상품을 찾기 위해 검색어를 입력하고 검색 결과를 보는 것 까지는 동일
    - 충성고객 코호트는 검색 결과에서 검색 필터 기능을 더 많이 사용
        - 검색 필터를 이용해 원하는 가격대, 색상, 디자인, 브랜드로 검색 결과를 한 번 더 좁히고
            
            $\rightarrow$ 그 결과에서 원하는 상품을 탐색하고 일반 고객보다 2~3개의 상품을 더 탐색 후 결제까지 이어지는 것을 발견

            $\rightarrow$ **검색 필터 사용이 결제 전환의 중요 티핑 포인트가 된다는 가설 수립 가능**

            $\rightarrow$ **일반 고객들에게도 검색 후 필터 사용에 대한 넛징이나 개입을 통해 실험하거나, UI/UX 자체를 개선하는 것**

## 저항 제거

<br>

### NUX 활용
- 신규 유저만을 위한 온보딩 사용 경험 플로우 $\rightarrow$ 신규 고객들이 빠르게 핵심 가치를 느끼게 해줄 수 있는 경험 설계
    - 예) 게임 튜토리얼
    - 예) SNS는 얼마나 빠르게 관심이 가고 취향이 맞는 계정을 팔로우 할 수 있느냐가 중요한 아하 모먼트 $\rightarrow$ 취향에 맞는 계정을 소개 하는 사용자 경험

<br>

### 긍정적인 저항
- 어떠한 저항은 제품이나 서비스의 신뢰도를 높일 수 있음
    - 예) 택시 호출 앱에서 택시를 불렀을 때 : '1km내에서 사용자를 찾고 있습니다'라는 안내 메시지

    $\rightarrow$ 실제 거리를 넓혀가며 탐색하고 있는 것을 보여주면, 현재 택시를 잡고 있다는 인식을 주며, 사용자로 하여금 인내심을 줄 수 있음
