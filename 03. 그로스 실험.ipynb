## AB테스트

<br>

### 좋은 AB테스트
- 가설 검증을 위한 세밀한 설계
- 왜 하나의 선택지가 더 좋은 결과를 얻었는지 쉽게 추론하고 이해

<br>

### 안 좋은 AB테스트
- 설계 없이 무작위로 진행
- 변수간의 상관관계를 이해하기 어려움 $\rightarrow$ 차이의 원인을 파악하기 힘듦

### 그로스 실험에 대한 한계

<br>

#### 완벽하게 통제되지 않는 실험 환경
- 통제 변수를 완벽히 통제학 힘듬
- 독립 변수 외에도 너무 많은 변수가 영향을 미침
- 실험/대조 그룹을 동일 조건으로 모집할 수 없음

<br>

### 그로스 마케터에게 필요한 실험
- **최소 요건**으로 가설 검증을 시작하여
- **연속적으로 일관된** 실험을 통해
- 고객 행동에 대한 **합리적인 인과 / 상관관계를 추론**하기 위한 실험이 필요

## 그로스 실험 문서
- **실험** : 검증을 목적으로 하는 소규모 실행
- **실행** : 실험에서 검증된 수단을 확장

<img src='img/0504.png' width=600>

<br>


### 그로스 실험 문서의 필요성
- 그로스 실험 문서를 통해
    - 스스로 실험 설계를 점검
    - 동료들에게 실험 방향성을 효과적으로 전달
    - 객관적인 실험 결과 해석

    $\rightarrow$ **실험을 통해 가설 검증**
