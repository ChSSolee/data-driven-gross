## Gross Framework : '성장'을 만들기 위한 접근법

<br>

#### 그로스 마케팅의 다양한 정의
- 테크 기업, 기술 중심 기업의 마케팅 전술?
- 제품 혹은 기술 기반으로 바이럴을 만드는 마케팅?
-  지출을 최소화하고 고객을 획득하는 마케팅 전략 및 활동?
- 데이터 분석이 고도화된 퍼포먼스 마케팅

    $\rightarrow$ **위는 그로스 마케팅에 대한 부분적인 설명**

<br>

#### 그로스 마케팅의 정의
- **사업이 성장(Growth)하기 위한 고객의 행동을 정량화하고, 이를 측정 가능한 방식으로 바꿔나가는 일**

    - **그로스마케팅은 회사가 필요로 하는 사업 성장 (Growth)을 만들기 위한 것**

    $\rightarrow$ **이를 위해 고객 행동에 집중하여 이 과정에서 고객의 행동을 원하는대로 구체화하고 정량화**

    $\rightarrow$ **정량화못하면 측정이 불가능하고, 측정이 불가능하면 관리할 수 없음**

    $\rightarrow$ **고객 행동의 변화는 측정 가능해야 함**

<br>

### 고객 행동의 정량화

1. 예) MAU

| 주어진 목표 | 행동 정량화 |
| - | - |
| **MAU 10만명 증가** | 1. **신규 고객** 10만명 획득<br>2. **기존 회원** 10만명 이탈 방지<br>3. **신규 고객** 5만명 획득 + **기존 회원** 5만명 이탈 방지 |

<br>

2. 예) 매출

| 주어진 목표 | 행동 정량화 |
| - | - |
| **매출 1천만원 달성** <br> (상품 가격 1만원) | 1. **신규 고객** 1,000명 확보하여 1개씩 구매<br>2. **기존 고객** 500명이 2개 구매<br>3. **신규 고객** 500명 확보 + **기존 회원** 250명이 2개 구매 |

<br>

3. 예) ROAS

| 주어진 목표 | 행동 정량화 |
| - | - |
| **ROAS 20% 개선** | 1. 같은 비용으로 20% 비싼 상품 판매<br> = 고객들이 **가격 인상**을 기꺼이 수용하게 만든다<br><br>2. 20% 저렴한 비용으로 고객 획득 <br> = 우리 제품을 **더 원하는 고객** 세그먼트를 발굴 <br><br>3. 우리 제품을 더 원하는 고객을 찾고 + <br> 더 비싼 상품을 구매하게 만듬 |

<br>

#### 그로스 마케터가 해야 하는 일
- **사업의 성장 (그로스)이라는 목표를 달성하기 위해 고객에게 어떤 행동을 요구할지, 어떤 행동을 이끌어낼지 구체적으로 정의하는 것이 중요**

<br>

- 예) 매출

| 행동 정량화 | 고객 행동 목표 수립<br>= 그로스 마케터가 해야하는 일 | 
| - | - |
| **신규 고객** 1,000명 확보하여 1개씩 구매 | **잠재고객 1만명**에게 도달해 강점 A를 살려 <br> **구매전환율 10%** 를 달성 | 
|**기존 고객** 500명이 이번달에는 2개 구매 | **기존 고객 1.5천명**에게 재도달해 강점 B를 살려 <br> **구매전환율 33%** 를 달성 |

<br>

- 예) ROAS

| 행동 정량화 | 고객 행동 목표 수립<br>= 그로스 마케터가 해야하는 일 | 
| - | - |
| 고객들이 **가격 인상**을 <br> 기꺼이 수용하게 만듬 | **구매자 후기 조회율**과 **쿠폰 사용률**을 높여 <br> 고단가 상품 **구매 전환율을 20%** 높임 | 
| 우리 제품을 **더 원하는** <br> **고객** 세그먼트를 발굴  | 광고 노출 빈도는 줄이고 도달 고객 범위는 넓혀 <br> **1, 2번의 노출로도 클릭율이 높은 고객**을 탐색 |

## 측정 가능하게 고객 행동 바꾸기

<br>

#### 실험/검증의 구조
- **모든 실험은 단번에 끝나지 않으며, 고객의 행동을 바꾸기 위한 실험은 연속적이고 체계적**

<br>


- 예) **B**라는 목표

1. **실험 1**
- 가설 : A $\rightarrow$ B
- 검증 : A $\rightarrow$ C

    $\rightarrow$ **A가 아닌 다른 가설 수립**

    $\rightarrow$ **D를 세분화하여 다시 가설 수립이후, 세분화 가설에 대한 반복**

### 측정 가능하게 고객 행동 바꾸기
- **첫 번째 실험에서 작은 단서로 작은 결과를 얻었다면, 두 번째 실험에서는 해당 단서를 구체화하여 조금 더 큰 결과를 검증**

    $\rightarrow$ **반복적인 실험을 통해, 확정적으로 특정 고객 행동을 이끌어내는 구체적인 수단 발견**

    $\rightarrow$ **구체적인 수단을 찾았다면, 해당 수단에 가진 리소스를 집중하여 큰 결과를 얻어냄**

    $\rightarrow$ **이러한 방식이 Agile (혹은 Lean Approach)**

#### Agile & Lean Approach
- 신속한 가설의 검증과 효율적인 성장
- **Lean Approach의 핵심은 가설과 이를 검증할 수 있는 효율적인 실험을 통해 빠르게 사업의 성장을 만들어낼 수 있는 원인과 결과를 파악하는 것**

    $\rightarrow$ **Lean Approach에서 그로스 마케팅이 도출되었으며, 그로스 마케팅은 고객의 행동을 측정 가능한 방식으로 바꿔나가는 데 같은 방법을 적용**

<img src='img/0005.jpg' width=400>

<br>

#### Agile vs WaterFall
- **WaterFall** : **업무의 단계별로 높은 완성도를 추구하여 최종 결과물 도출**
    - 예) 고객 조사 $\rightarrow$ 기획 $\rightarrow$ 디자인 $\rightarrow$ 실행

<img src='img/0006.jpg' width=400>

<br>

- **Agile은 큰 일을 작은 단위로 쪼개서 연속되는 가설의 수립과 실험을 통해 접진적으로 조금씩 큰 성과 (여러번 나눠서 검증)**

    $\rightarrow$ **WaterFall은 한 번에 큰 성과**

- Agile은 처음 기획이나 가설의 방향이 중간에 바뀌는 것이 당연하고 자연스러움

<img src='img/0007.jpg' width=600>

<br>

- 고객 행동을 측정 가능한 방식으로 바꿔나가야 하는 그로스 마케팅에세 Agile 접근법은 필수

    $\rightarrow$ **약간의 불확실함과 불완전함을 감수하고서라도, 빠르게 고객의 반응을 확인하여 교훈을 얻고, 이 교휸을 기획에 반영하여 지속적인 개념을 효율적으로 만들어 나가는 것이 Lean Approach의 핵심**

## 고객 행동 퍼널
-  **Input과 Output이 연쇄적으로 연결된 퍼널에서, 최종 Output이 늘어나기 위해서는 이전 단계의 Input이 늘어나거나 각 단계별 전환율이 높아져야 함**

    $\rightarrow$ 고객은 우리 제품과 서비스에 대한 여러 가지 행동을 연쇄적으로 시행

    $\rightarrow$ 이를 Input, 전환, Output의 관계로 도식화 $\rightarrow$ 고객 행동 퍼널

<br>

- 마케터는 고객에게 원하는 최종 Output 행동을 지정하고, 이를 원하는 만큼 늘리거나 줄이는 일을 함

    $\rightarrow$ 마케터가 원하는 Output 행동을 늘리거나 줄이기 위해서는, Input과 전환을 늘리거나 줄임

    $\rightarrow$ 마케터는 제품과 서비스에  대한 고객 행동 퍼널을 체계화 하고, 이를 구성하는 Input과 Output의 관계를 파악하여 마케터가 원하는 방식으로 Output을 만들어 나가는 일

### AARRR
- 일반적인 고객의 행동을 구조화 한 것

    $\rightarrow$ 모든 서비스, 제품, 사업에 그대로 적용되기 힘들 수 도 있음

    $\rightarrow$ **어떤 서비스나 제품에도 적용되는 퍼널을 제시하기 보다는, 고객 행동 퍼널을 중심으로 고객의 행동을 분석하고, 문제를 찾고, 해결할 수 있다는 관점을 보여주는 것**

<br>

#### 01. Acquisition : **고객이 특정 제품이나 서비스로 유입**
- 측정 지표

| **지표** | **설명** | 
| - | - |
| **인지도** | 전체 고객 중 우리 서비스나 제품을 알고 있는 고객의 비율 |
| **TOM** | 전체 고객 중 특정 카테고리에서 우리 서비스나 제품을 '가장 먼저 떠올리는' 고객의 비율 |
| **노출, 조회, 도달, 빈도** | 노출 or 조회 = 도달 X 빈도 |
| **Share of Voice (SoV)** | 관련 키워드 언급 콘텐츠에서 자사에 대한 콘텐츠의 노출 비율 |
| **Share of Search (SoS)** | 관련 키워드 검색 결과에서 자사에 대한 콘텐츠의 노출 비율 |
| **Traffic (UV, PV)** | UV : 유니크 방문자 수 <br> PV : 페이지 조회 수 |
| **CPM** | 1천회 노출 당 마케팅 비용 (=노출수 / 마케팅 예산 * 1,000) |
| **CPC** | 1회의 조회 당 마케팅 비용 / 1회의 클릭 당 마케팅 비용 |
| **CAC, CPA** | 1명의 획득 당 마케팅 비용 |

<br>

#### 02. Activation : 해당 서비스가, 고객이 가지고 있는 니즈를 해소해줄 수 있는지를 알아보고 탐색
- 웹 사이트나 앱을 이리저리 살펴보면서 어떤 상품이 있는지, 어떤 혜택을 제공하는지 발견하고 이해하는 과정
- 측정 지표

| **지표** | **설명** | 
| - | - |
| **MAU, WAU, DAU** | 월간 활성 유저, 주간 활성 유저, 일간 활성 유저 |
| **Engagement, Stickiness** | MAU중 DAU의 비율 (=DAU / MAU) |
| **앱설치, 회원가입** | 단순 방문 이상의 고객의 참여가 필요한 앱설치나 회원가입한 고객의 수 |
| **Traffic (UV, PV)** | UV : 유니크 방문자 수 <br> PV : 페이지 조회 수 |
| **진입율, 사용률, 전환율** | 전체 방문자 중 특정 페이지에 진입/사용/구매한 고객 비율 |
| **CTR** | 노출된 사람들 중 클릭한 사람의 비율 |


<br>

#### 03. Revenue : 고객이 지갑을 여는 단계
- 고객이 가진 니즈가 해소될 수 있는 핵심 가치를 발견하여, 고객이 결제하는 행동
- 측정 지표 : 고객의 결제 행위와 관련된 지표

| **지표** | **설명** | 
| - | - |
| **GMP (Gross Merchandise Volume)** | 거래액 |
| **AOV** | 주문당 평균 결제 금액 |
| **ASP** | 고객 당 결제 상품의 평균 가격 |
| **ARPU** | **전체 고객** 당 매출 |
| **ARPPU** | **구매 고객** 당 매출 |
| **AMPU** | 활성 고객 당 수익 (**마진**) |
| **구매 전환율, 결제 전환율** | 전체 방문자 중 구매/결제한 고객 비율 |
| **교차구매율** | 구매자 중 타 카테고리로 교차 구매한 고객 비율 |
| **취소율, 반품율** | 구매한 고객 중 취소/반품한 고객의 비율 |

<br>

#### 04. Retention : 결제한 고객이 해당 서비스나 상품을 지속 사용하면서 충성도가 높아지는 단계
- 첫 결제를 하고 만족도가 높아진 고객이 웹사이트나 앱을 재방문하고 재사용률이나 재구매율이 높아지는 과정
- 측정 지표

| **지표** | **설명** | 
| - | - |
| **재방문율, 재사용율** | 방문한 고객이 다시 방문하는 비율, 사용한 고객이 다시 사용하는 비율 |
| **재구매율, 재결제율** | 구매한 고객이 재구매하는 비율, 결제한 고객이 재결제하는 비율 |
| **이탈율 (Churn)** | 방문/사용/구매한 고객이 다시 방문/사용/구매하지 않는 비율 |
| **복귀율** | 이탈한 고객이 다시 방문/사용/구매하는 비율 |
| **LTV** | 고객 생애 가치 |
| **LTR** | 고객 생애 매출 |
| **앱 삭제, 회원탈퇴** | 앱을 삭제하는 고객의 수, 회원을 탈퇴하는 고객의 수 |

<br>

#### 05. Referral : 충성도와 만족도가 높아진 고객이 해당 제품이나 서비스를 지인들에게 추천하는 단계
- 높은 수준의 충성도를 의미하며, 이 퍼널까지 도달한 고객은 제품이나 서비스의 홍보대사와 같은 역할

    고객이 새로운 고객을 불러오는 상태
- 측정 지표

| **지표** | **설명** | 
| - | - |
| **공유** | 한 명의 고객이 만들어 내는 공유 횟수 |
| **리퍼럴** | 한 명의 고객이 만들어 내는 신규 고객 획득 수 |
| **Word of Mouth (WoM)** | 고객이 만들어 내는 자사 관련 콘텐츠 수 혹은 언급량 |
| **NPS (순수 추천 지수)** | 고객 중 적극적으로 해당 서비스를 추천하는 고객의 비율 |
| **리뷰 스코어** | 앱스토어, 판매 채널 등에서 구매 고객들이 남기는 후기의 점수 |
| **바이럴 계수** | 고객 한 명이 만들어 낼 수 있는 바이럴의 크기 |



<br>




### AARRR 퍼널 도입
- AARRR 퍼널을 도입하는 것은, 퍼널의 순서 그대로를 적용하는 것이 아닌,

    **고객 행동을 퍼널로 정의하고 이를 통해 개선점을 찾는 것**
- AARRR퍼널의 각 단계별 주요 지표를, 그로스 마케팅 조직의 목표나 핵심 KPI로 설정

<br>

#### 퍼널별 중요도
- 예) 쿠팡과 같은 커머스 플랫폼은 앱 진입 직후부터 탐색이 시작

    $\rightarrow$ 잦은 재방문과 재구매가 발생
