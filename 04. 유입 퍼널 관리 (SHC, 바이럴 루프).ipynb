## 고객 행동 퍼널

<br>

#### 오프 사이트 퍼널 (Off-Site)
- 잠재 고객들
- 우리 서비스로 유입되기 전에 경험하는 행동과 인식
- 서비스 외부에서 검색, 광고 노출 등

<br>

#### 온사이트 퍼널 (On-Site)
- 유입 고객
- 우리 서비스로 유입된 이후 경험하는 퍼널
- 앱 설치, 가입, 탐색, 구매 AARRR

#### 인지 (Awareness)

| **지표** | **설명** | 
| - | - |
| **인지도** | 전체 고객 중 우리 서비스나 제품을 알고 있는 고객의 비율 |
| **TOM** | 전체 고객 중 특정 카테고리에서 우리 서비스나 제품을 '가장 먼저 떠올리는' 고객의 비율 |
| **노출, 조회, 도달, 빈도** | 노출 or 조회 = 도달 X 빈도 |
| **검색 점유율** | 경쟁사를 포함한 특정 카테고리의 검색량 중 자사의 키워드 검색량의 점유율 |
| **검색 결과 점유율** | 관련 키워드 검색 시 검색 결과에서 자사에 대한 콘텐츠가 얼마나 노출되는지 |
| **언급량** | 소셜 미디어, 온라인 커뮤니티 등에서 자사에 대해 언급하는 포스트/콘텐츠의 양 |
| **CPM** | 1천회 노출 당 마케팅 비용 |
| **CPR** | 1명 고객 도달 당 마케팅 비용 |
| **CPV** | 1회 조회 당 마케팅 비용 |

## 지속 가능한 유입 전략

<br>

### 유입 경로 최적화

<img src='img/0607.png' width=600>

#### Paid 경로
- 유료로 고객을 획득하는 채널
- 모든 형태의 광고 (DA, SA, TVC 등)
- 협찬, PPL
- 제휴, Affiliate 프로그램

<br>

#### Organic 경로 (=Unknown)
- 자연적으로 고객을 획득하는 채널 (식별 불가 경로)
- 검색, 소셜 미디어, 유튜브, 방송 노출
- 블로그, 커뮤니티, 카페, 뉴스, 잡지
- 플랫폼 교차 사용 (백링크)

<br>

#### Viral 경로
- 유입 단위 당 비용이 발생하지 않거나, 발생 하더라도 유입 전파성이 높은 경로
- 조회수가 터진 유튜브 영상
- 유명 채널에 소개된 팝업 스토어
- ... 

<br>

### 유입 경로 최적화
- 획득 고객들의 **기여 유입 경로** 파악
- 잠재 고객들의 유입 전 **오프사이트 퍼널 행동** 확인

<br>

#### 획득 고객들의 기여 유입 경로

<img src='img/0608.png' width=600>

- **기여 기간 (룩백 윈도우)** : 고객 획득 전 몇일 까지를 유효 기여 기간으로 볼 것인가?
    -  1, 3, 7일 등
- **기여 비중** : 기여 기간 중 발생한 기여 이벤트들의 기여 비중을 어떻게 측정할 것인가?
    - 첫 클릭, 마지막 클릭, 분배 등
- **기여 방식** : 기여 기간 중 발생한 이벤트들의 기여 조건을 어떻게 정의할 것인가?
    - 클릭 이벤트, 조회 이벤트 등

<br>

#### 잠재 고객들의 오프사이트 퍼널 행동

| **체크 리스트** | **관련 유입 채널** |
| - | - |
| 잠재 고객들이 문제를 해결하기 위해 검색을 하는가? | 검색 최적화, 검색 광고 |
| 충동 구매나 비계획 구매가 일어날 수 있는 상품/서비스인가? | 소셜 미디어 광고 |
| 구매 빈도가 높은 상품/서비스인가? | 소셜 미디어 광고 |
| 지인에게 추천하거나 추천 받을 명분이 있는 상품/서비스인가? | 레퍼럴, 추천 프로그램|
| 유저가 늘어날수록 고객의 편의성이나 가치도 함께 늘어나는 상품/서비스인가 ? | 레퍼럴 추천 프로그램 |
| 구매 확신을 가지는데 공신력이나 검증이 필요한 상품/서비스인가? | 홍보, PR, 제휴|
| 많은 대안이 존재하여 큐레이션이나 추천이 필요한 상품/서비스인가? | 인플루언서, 콘텐츠 |
| 심도 있는 설명이나 사용법에 대한 숙지가 필요한 상품/서비스인가? | 인플루언서, 콘텐츠 | 
| 강한 경쟁사나 선도 경쟁사가 존재하는가 ? | 유료 광고 |
| 높은 LTV를 기대할 수 있는 상품이나 서비스인가? | 유료 광고 |

## 바이럴 루프

<br>

### 바이럴 계수

> **바이럴 계수 = 기존 유저의 도달 범위 $\times$ 전환율 $\times$ 빈도**

- **기존 유저의 도달 범위** : 기존 고객이 한 번의 액션으로 몇 명의 사람에게 우리 서비스를 알려 줄 수 있는지
- **전환율** : 기존 고객에 의해 도달된 잠재 고객들 중 실제 우리 서비스에 유입된 고객의 비율
- **빈도** : 기존 고객이 얼마나 자주 지인들에게 우리 서비스를 추천하는지

<br>

### 바이럴 루프 메커니즘

<br>

#### 이중 유인책
- 초대하는 사람과 초대를 받는 살마 모두가 혜택을 받을 수 있도록 인센티브 설계

<br>

#### 자체 전파성 활용
- 제품이나 서비스 자체가 가지는 전파성이나 네트워크 효과를 활용
    - 토스의 '송금하기', 카카오톡의 '선물하기' 등

<br>

#### 전파성 높은 이슈
- 기존 고객들과 외부의 잠재 고객들이 관심을 가지고 공유될 만한 전파성이 높은 이슈를 만들어 바이럴 유도
    - 토스의 주식 한주 무료 증정 이벤트 등

<br>

### 바이럴 루프 예산
- **바이럴 루프를 통해 발생하는 단위당 고객 획득 비용이, 기존 CAC보다 저렴하다면**

    $\rightarrow$ 지속 가능한 바이럴 루프 예산

    - 바이럴 루프의 인센티브 비요잉 CAC보다 커지게 되면 굳이 바이럴 루프를 진행할 필요가 없음

        $\rightarrow$ 같은 비용으로 기존 마케팅 활동을 유지하는것이 더 효율적이기 때문

    - **CAC는 제품/서비스 마다 모두 다르기에, 바이럴 루프에 필요한 예산 책정 시, CAC에 집중**
