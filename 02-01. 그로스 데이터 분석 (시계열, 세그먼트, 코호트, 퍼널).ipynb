## 그로스 마케터에게 필요한 데이터
- 그로스 마케터에게 데이터는 필수적
- 고객의 행동을 정량화하고, 이를 측정해 나가기 위해서 데이터가 반드시 필요하며, 마케터는 이를 기반으로 가설을 수립하여 필요한 실험을 지진행


<img src='img/0301.png' width=600>

<br>

- **고객이 어떠한 행동을 하게 되면 두 가지의 데이터가 생성**
    - **액션** : 실제로 고객이 어떤 행동을 했는가?
    - **속성** : 고객의 액션을 서ㅔ부적으로 이해할 수 있는 특징

<img src='img/0302.png' width=600>

<img src='img/0306.png' width=300>

<br>

- 속성의 유형
    - **고객 속성** : 성별 연령, 가입 일 등 **고정적**
    - **행동 속성** : 상품 명, 상품 개수 등 행동에 따라 **가변적**

<img src='img/0307.png' width=600>

<br>

### 마케터의 데이터 분석

<br>

#### 분석하려는 액션의 선택
- **행동에 대한 액션과 함께 액션에 연결된 행동 속성, 고객에게 연결된 고객 속성이 만들어짐**

    $\rightarrow$ **고객 속성에서는 진입 경로, 고객 프로파일이 중요**

- 해당 행동이 고객 속성과 행동 속성을 파악했다면, 그 행동 전/후로 발생한 다른 행동도 파악

    **전/후로 행한 고객의 액션을 잘 살펴보면, 상관성과 행동의 의도를 유추할 수 있는 경우가 많기 때문**

    $\rightarrow$ **마케터는 특정 고객의 행동을 분석하기 위해 해당 행동의 고객 속성과 행동 속성을 파악하고,** 
    
    **전/후 고객의 연결된 행동을 보면서 특정 고객이 보이는 상관성과 행동 의도를 파악해, 왜 고객이 이런 행동을 했는지 추론**

### 시계열 분석
- 분석하고 싶은 특정 속성의 고객이나 특정 고객의 행동이 있다면, 시간의 흐름에 따른 그 값의 변화를 살펴보는 것
- 시계열 분석의 주요 목적
    1. **변곡점 탐색** : 변곡점을 탐색하면, 그 변곡점에 유효하게 작용한 특정 변수가 고객의 행동에 영향을 미쳤을 가능성이 높음

        $\rightarrow$ 무엇이 그 변곡점에 영향을 주었는가 탐색

    2. **상관관계 탐색** : 시계열에서 만들어지는 그래프가 비례 혹은 반비례 관계가 보인다면, 두 요소가 상관관계를 가질 수도 있음

### 세그먼트 분석
- **내가 원하는 행동을 보이는 고객과, 그렇지 않은 고객을 비교하여, 각 고객의 특성 분석**

<img src='img/0310.png' width=600>

<br>

#### 공통점과 차이점 탐색
- 내가 원하는 행동을 수행한 고객과 그렇지 않은 고객을 구분하고,

    고객의 속성부터 행동의 속성까지 무엇이 같고 무엇이 다른지를 파악

    - 예) 구매 고객 vs 미구매 고객
        - 고객의 연령, 성별, 유입 경로, 가입일 등 다양한 고객 속성과

            고객 상품, 구매 상품의 가격대 등의 거래 속성을 비교

<br>

#### 정도의 차이를 탐색
- 내가 원하는 행동을 많이 한 고객부터 적게 한 고객까지 그 정도를 나누고, 정도가 높고 낮은 고객들 사이에 차이점이 있는지 살펴보는 것

    - 예) 한 달 동안 구매라는 행동을 많이 한 고객과 적게한 고객을 분류
        - 분류한 각 고객들의 고객이나 구매 속성을 탐색
        - **내가 원하는 행동을 더 많이 하는 고객을 분석해, 무엇이 그 행동을 유발하는지 다른 세그먼트와 비교**

<br>

### 코호트 분석
- 특정 행동을 한 고객들을 지정해 놓고, 그 고객이 과거에 무슨 일을 했는지

    혹은 반대로 시간이 경과한 후에 그 시점 이후 무슨일을 했는지 탐색 (시간 추적)


<br>

### 퍼널 분석
- 고객의 행동을 퍼널 형태로 분석
- **코호트가 고객을 중심으로 고정 vs 퍼널은 행동과 액션 기준으로 고정**
    - **코호트가 특정 행동을 한 고객을 지정해 놓고, 그 고객의 과거와 미래를 탐색**
    - **퍼널 분석은 특정 행동을 지정하고, 그 행동을 한 고객들이 어떻게 변화하는지를 탐색**


<br>

### 코호트 + 퍼널 (하이브리드)
- **서로 다른 행동을 하는 고객들의 세그먼트를 나누고, 세그먼트별로 퍼널 분석이나 코호트 분석**

    고객 세그먼트별로 각각의 퍼널을 구조화해 차이를 비교
    
    시간이 경과 함에 따라 어떻게 행동이 바뀌는지를 탐색

<img src='img/0311.png' width=800>

<br>

#### 분석 사례
- **가입율** : 웹 사이트에 방문한 고객들 중 특정 기간 동안 회원가입까지 완료한 고객들의 비율 (**시간 기준에 따라 가변**)
- N일 이내에 상세페이지에 얼마나 들어갔는지, 상세페이지에 들어간 고객들이 N일 이내 얼마나 구매 하는지 등을 코호트 개념으로 탐색
    
    재구매율은 조금 더 긴 시간의 관점 (30일 등)

<br>

- **자연 유입 고객** :
    - 구매율이 낮은 반면, 재구매율이 높음 $\rightarrow$ 많이 확보될 수록 미래의 구매까지 키울 수 있는 고객
    - **상세페이지 진입율이 낮음 $\rightarrow$ 맞춤형 상품을 추천하여, 상품 탐색의 기회를 준다면 성장 가능**
- **검색 광고 유입 고객**
    - 재구매율은 매우 낮지만, 가입 후 구매까지의 행동이 매우 빠르며 전환율이 높음 $\rightarrow$ 첫 구매 이후 재구매 유도를 위한 적극적인 CRM 활동 필요
- **콘텐츠 유입 고객**
    - 재구매율과 구매율이 높으나, 가입율이 낮음 $\rightarrow$ 유입 직후 빠르게 가입을 완료할 수 있도록 인센티브 고려 가능
